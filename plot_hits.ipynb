```python
import matplotlib.pyplot as plt


def read_hits(name, ax1, ax2):
    file = open(name + ".txt", "r")
    content = file.readlines()
    file.close()

    x, y1, y10 = [], [], []
    for i in range(len(content)):
        x.append(i)
        y1.append(float(content[i].split()[0]))
        y10.append(float(content[i].split()[1]))

    # Plot Hits@1 in ax1
    ax1.plot(x, y1, label="Hits@1 " + name)
    
    # Plot Hits@10 in ax2
    ax2.plot(x, y10, label="Hits@10 " + name)


# Create two subplots (stacked vertically)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

# Read and plot data
names = [
    "hit_train_seq_22_80_20_contrast_quad_hilbert_w100_1_4_cosine_wr",
    "hit_train_seq_22_80_20_contrast_quad_hilbert_w100_1_4_cosine",
    "hit_train_seq_22_80_20_contrast_quad_hilbert_w1000_5_5_cosine",
    "hit_train_seq_22_80_20_contrast_quad_hilbert_n3_lr5_5",
    "hit_train_seq_22_80_20_contrast_quad_hilbert_w2000_5_5",
    "hit_train_seq_22_80_20_contrast_quad_hilbert_w100_5_4_cosine",
    "hit_train_seq_22_80_20_contrast_quad_hilbert"
]

for name in names:
    read_hits(name, ax1, ax2)

# Formatting subplot 1 (Hits@1)
ax1.set_title("Hits@1 score during validation")
ax1.set_xlabel("Iteration x 100")
ax1.set_ylabel("Hits@1 Score")
ax1.legend()
ax1.grid()

# Formatting subplot 2 (Hits@10)
ax2.set_title("Hits@10 score during validation")
ax2.set_xlabel("Iteration x 100")
ax2.set_ylabel("Hits@10 Score")
ax2.legend()
ax2.grid()

# Adjust layout
plt.tight_layout()
plt.show()
```
