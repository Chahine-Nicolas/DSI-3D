```python
import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Required for 3D plotting
from module_loader_kitti_pose import *  # Add for more metrics
from hilbertcurve.hilbertcurve import HilbertCurve
import numpy as np
import json

##################################
# Read pose data
##################################
eval_dataset = 'Kitti'
eval_seq = 6
kitti_dir = "/lustre/fsn1/worksf/projects/rech/dki/ujo91el/datas/datasets/"
eval_seq = '%02d' % eval_seq
sequence_path = kitti_dir + 'sequences/' + eval_seq + '/'

# Load poses
tfs, pose = load_poses_from_txt(sequence_path + 'poses.txt')

# Normalize pose by subtracting the minimum bounding box
min_bbox = np.min(pose, axis=0)
pose = pose - min_bbox

# Define Hilbert curve parameters
p = 17  # Number of iterations (depth of the Hilbert curve)
n = 2   # Number of dimensions (2D)
hilbert_curve = HilbertCurve(p, n)

# Hilbert
# Prepare data for plotting
x_vals = []
y_vals = []
z_vals = []
colors = []

for i in range(len(pose)):
    label_id_gps = pose[i]
    xx = round(label_id_gps[0] * 100)
    yy = round(label_id_gps[1] * 100)

    # Calculate the Hilbert curve distance
    res_str = hilbert_curve.distances_from_points([[xx, yy]])[0]

    # Collect data for the 3D plot
    x_vals.append(label_id_gps[0])
    y_vals.append(label_id_gps[1])
    z_vals.append(res_str)
    colors.append(i)  # Use the index `i` to progressively change color

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data
scatter = ax.scatter(x_vals, y_vals, z_vals, c=colors, cmap='jet', marker='o')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Hilbert Curve index')

# Show the plot
plt.show()


# GPS
# Prepare data for plotting
x_vals = []
y_vals = []
z_vals = []
colors = []

for i in range(len(pose)):
    label_id_gps = pose[i]
    xx = round(label_id_gps[0] * 100)
    yy = round(label_id_gps[1] * 100)

    # Calculate res_str as a concatenated string
    xx_str = f'{xx:05}'
    yy_str = f'{yy:05}'
    res_str = ''.join(x + y for x, y in zip(xx_str, yy_str))
    res_str += xx_str[len(yy_str):] + yy_str[len(xx_str):]

    # Collect data for the 3D plot
    x_vals.append(label_id_gps[0])
    y_vals.append(label_id_gps[1])
    z_vals.append(int(res_str))  # Convert res_str to integer for plotting
    colors.append(i)  # Use the index `i` to progressively change color


plt.figure()
plt.scatter(x_vals, y_vals, c=colors, cmap='jet', marker='o')

# Set 2D plot labels
plt.xlabel('X')
plt.ylabel('Y')

# Show the plots
plt.show()

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data
scatter = ax.scatter(x_vals, y_vals, z_vals, c=colors, cmap='jet', marker='o')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('GPS index')

# Show the plot
plt.show()


# hierarchical
# Prepare data for plotting
x_vals = []
y_vals = []
z_vals = []
colors = []

f = open(sequence_path+"hierarchical.json",) # gps xyxyxyxy
data = json.load(f)
f.close()

for i in range(len(pose)):
    label_id_gps = pose[i]
    res_str = data[str(i)]

    # Collect data for the 3D plot
    x_vals.append(label_id_gps[0])
    y_vals.append(label_id_gps[1])
    z_vals.append(int(res_str))  # Convert res_str to integer for plotting
    colors.append(i)  # Use the index `i` to progressively change color

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data
scatter = ax.scatter(x_vals, y_vals, z_vals, c=colors, cmap='jet', marker='o')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Hierarchical index')

# Show the plot
plt.show()

# label
# Prepare data for plotting
x_vals = []
y_vals = []
z_vals = []
colors = []

f = open(sequence_path+"hierarchical.json",) # gps xyxyxyxy
data = json.load(f)
f.close()

for i in range(len(pose)):
    label_id_gps = pose[i]
    res_str = data[str(i)]

    # Collect data for the 3D plot
    x_vals.append(label_id_gps[0])
    y_vals.append(label_id_gps[1])
    z_vals.append(int(i))  # Convert res_str to integer for plotting
    colors.append(i)  # Use the index `i` to progressively change color

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data
scatter = ax.scatter(x_vals, y_vals, z_vals, c=colors, cmap='jet', marker='o')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Label index')

# Show the plot
plt.show()


    
```
