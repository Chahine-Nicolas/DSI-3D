```python
###############################
eval_seq = 0
file_closest =  open('/gpfswork/rech/dki/ujo91el/code/logg3dnet/closest_seq_'+str(eval_seq)+".txt","w")
dictio = {}
start_time = timestamps[0]
for query_idx in range(num_queries):
    
    query_time = timestamps[query_idx]
    
    if is_revisit_list[query_idx] == 1.0:
        query_pose = positions_database[query_idx]
        
        d = []
        q = []
        
        for comp_idx in range(num_queries):
            comp_time = timestamps[comp_idx]
            
            if (abs(query_time - comp_time) - cfg.skip_time) < 0:
                    continue
            if query_idx != comp_idx:
                comp_pose = positions_database[comp_idx]
                p_dist = np.linalg.norm(query_pose - comp_pose)
                d.append(round(p_dist,3))
                q.append(round(comp_idx,3))

        d, q = zip(*sorted(zip(d, q)))
        print("query_idx  ", query_idx)
        print(d[:10])
        print(q[:10])
        file_closest.write(str(query_idx)+" dist :"+str(d[:10])+" query :"+str(q[:10])+"\n")
        dictio[query_idx] = [d[:10], q[:10]]
        
with open('closest_seq_'+str(eval_seq)+'.json', 'w', encoding ='utf8') as json_file: 
    json.dump(dictio, json_file, allow_nan=False) 
file_closest.close()
```
